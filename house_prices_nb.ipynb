```python
pp_cont_base_imp = SimpleImputer()
pp_cont_base_scaler = MinMaxScaler()
pp_cont_baseline = make_pipeline(pp_cont_base_imp, pp_cont_base_scaler)

pp_cat_base_imp = SimpleImputer(strategy='most_frequent')
pp_cat_base_encoder = OneHotEncoder(handle_unknown='ignore')
pp_cat_baseline = make_pipeline(pp_cat_base_imp, pp_cat_base_encoder)
                                
pp_baseline = make_column_transformer(
    (pp_cont_baseline, make_column_selector(dtype_exclude=["O"])),
        (pp_cat_baseline, make_column_selector(dtype_include=["O"])))
```

```python
pp_baseline
```

```python
pp_baseline.fit_transform(X).shape
```

```python
def root_mean_squared_log_error(y_true, y_pred):
    t = np.array(y_true)
    p = np.array(y_pred)
    log_error = np.log(1+t) - np.log(1+p)
    return ((log_error**2).mean())**0.5

def root_mean_squared_log_error_neg(y_true, y_pred):
    t = np.array(y_true)
    p = np.array(y_pred)
    log_error = np.log(1+t) - np.log(1+p)
    return -1 * ((log_error**2).mean())**0.5


rmsle = make_scorer(root_mean_squared_log_error)

rmsle_neg = make_scorer(root_mean_squared_log_error_neg)
```

```python
model_base = DecisionTreeRegressor()
pipe_base = make_pipeline(pp_baseline, model_base)
```

```python
pipe_base
```

```python
score_base = cross_val_score(pipe_base, X, y, cv=5, scoring=rmsle).mean()
score_base
```

```python
model_base_XGBR = XGBRegressor()
pipe_base_XGBR = make_pipeline(pp_baseline, model_base_XGBR)
score_base_XGBR = cross_val_score(pipe_base_XGBR, X, y, cv=5, scoring=rmsle).mean()
score_base_XGBR
```

```python
model_base_RF = RandomForestRegressor()
pipe_base_RF = make_pipeline(pp_baseline, model_base_RF)
score_base_RF = cross_val_score(pipe_base_RF, X, y, cv=5, scoring=rmsle).mean()
score_base_RF
```
